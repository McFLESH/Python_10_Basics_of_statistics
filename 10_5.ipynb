```python
# Вам требуется создать функцию website_analyse, которая принимает на вход путь к файлу с входной таблицей.
# В функции необходимо проанализировать таблицу и ответить на следующие вопросы в формате, указанном в конце задания:
# Какое среднее количество просмотренных страниц пользователями сайта во всей таблице?
# Какое стандартное отклонение времени, проведенного на сайте?
# Рассчитайте коэффициент вариации и сделайте вывод о том, насколько изменчиво поведение пользователей сайта в зависимости от времени нахождения на сайте.
# Проверьте, есть ли статистически значимая разница в количестве страниц, просмотренных пользователями с мобильных устройств и десктопов.



import pandas as pd
import numpy as np
from scipy import stats

def website_analyse(file_path):
    data = pd.read_csv(file_path)
    pages = round(np.mean(data["Просмотренные страницы"]),2)
    print(f'1) Среднее количество просмотренных страниц: {pages}')

    sec = round(np.std(data["Время на сайте (сек)"]), 2)
    print(f'2) Стандартное отклонение: {sec}')

    CV = round(np.std(data["Время на сайте (сек)"])/np.mean(data["Время на сайте (сек)"]), 2)
    if CV > 0.25:
      print(f'3) Высокая вариабельность: {CV}')
    elif 0.1 < CV < 0.25:
      print(f'3) Умеренная вариабельность: {CV}')
    else:
      print(f'3) Слабая вариабельность: {CV}')

    data1 = data[data['Тип устройства'] == 'Мобильное']
    data2 = data[data['Тип устройства'] == 'Десктоп']
    a = list(data1['Просмотренные страницы'])
    b = list(data2['Просмотренные страницы'])
    p_value = round(stats.ttest_ind(a, b)[1], 6)
    if p_value < 0.05:
      print(f'4) Есть статистически значимая разница в количестве страниц: {p_value}')
    else:
      print(f'4) Нет статистически значимой разницы: {p_value}')
```
