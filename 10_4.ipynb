```python
# В этом задании вам предстоит работать с файлом NZL_le.csv.
# Он представляет из себя датасет с данными о продолжительности жизни в Новой Зеландии в период с 1950 по 2021 год.

# Объявите функцию get_variation(), которая принимает переменную file_name -  название csv файла.
# Функция должна рассчитать коэффициент вариации по данным о мужской продолжительности жизни

import pandas as pd
import numpy as np

def get_variation(file_name):
    data = pd.read_csv(file_name)
    s = np.std(data["Male Life Expectancy"])
    k = np.mean(data["Male Life Expectancy"])
    CV = float(s/k*100)
    return round(CV, 2)
```
