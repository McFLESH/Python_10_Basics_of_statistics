```python
# Объявите функцию get_statistics(), которая принимает одну переменную int_list – список случайных целых чисел.
# Функция должна поcчитать среднее арифметическое и медиану заданного массива.

import pandas as pd

def get_statistics(int_list):
    s = pd.Series(int_list)
    a = [float(s.mean()), s.median()]
    return a
```
