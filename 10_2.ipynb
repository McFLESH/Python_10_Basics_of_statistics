```python
# Объявите функцию get_mode(), которая принимает одну переменную int_list – список случайных целых чисел.
#Функция должна найти моду и вернуть её в виде числа. Для решения задания воспользуйтесь библиотекой pandas.

import pandas as pd

def get_mode(int_list):
    s = pd.Series(int_list)
    a = s.mode()
    return a[0]
```
