```python
# Объявите функцию get_stats(), которая принимает две переменные – размер массива size и его моду mode.
# С помощью библиотеки numpy сгенерируйте внутри функции массив необходимого размера с заданной модой.
# Для данного массива посчитайте медиану, среднее арифметическое и стандартное отклонение.
# Занесите вычисления в список в следующем порядке: [медиана, среднее, стандартное отклонение].

import numpy as np

def get_stats(size, mode):
  arr = np.full(size, mode) + np.random.randn(size)
  b = [np.median(arr), np.mean(arr), np.std(arr)]
  return arr, b
```
